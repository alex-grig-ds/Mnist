## Errors analysis

```python
import math
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ERRORS_FILE = './errors.csv'
SAMPLES_QNTY = 16
```

```python
errors_df = pd.read_csv(ERRORS_FILE)
errors_df = errors_df.sample(SAMPLES_QNTY, random_state=123, ignore_index=True)

side_size = int(math.ceil(SAMPLES_QNTY**0.5))
plt.figure(figsize=(10,10))
for idx, sample in errors_df.iterrows():
    image = Image.open(sample['path'])
    plt.subplot(side_size, side_size, idx+1)
    plt.imshow(image)
    plt.title(f"Label: {sample['label']}. Predict: {sample['predict']}")
    plt.axis('off')

plt.show()
```
